```python
load = lambda x: pd.read_csv(
    x,
    index_col=0,
    header=0,
    parse_dates=["DATE"],
    date_parser=lambda d: datetime.strptime(d, "%Y-%m-%d")
)
# Load crude oil dataset
df = load("/Users/tianyudu/Documents/UToronto/Course/ECO499/ugthesis/data/ready_to_use/rpna.csv")
df_raw = pd.read_csv("../data/ravenpack/crude_oil_all.csv")
```

```python
def plot_ts(columns: Union[str, List[str]]):
    if not isinstance(columns, list):
        columns = [columns]
    for col in columns:
        data = go.Scatter(
            x=df.index,
            y=df[[col]].values.squeeze(),
            line=go.scatter.Line(color="blue", width=0.8),
            opacity=0.8,
            name=col,
            text=[f"{col}: {x:.0f}" for x in df[col].values]
        )
        layout = go.Layout(
            title=col,
            xaxis=dict(
                title='Date',
                # Range selector with buttons
                rangeselector=dict(
                    # Buttons for selecting time scale
                    buttons=list([
                        # 10 year
                        dict(count=10,
                          label='10y',
                          step='year',
                          stepmode='backward'),
                        # 3 year
                        dict(count=3,
                          label='3y',
                          step='year',
                          stepmode='backward'),
                        # 1 year
                        dict(count=1,
                          label='1y',
                          step='year',
                          stepmode='backward'),
                        # 6 months
                        dict(count=6,
                          label='6m',
                          step='month',
                          stepmode='backward'),
                        # 1 month
                        dict(count=1,
                          label='1m',
                          step='month',
                          stepmode='backward'),
                        # 1 week
                        dict(count=7,
                          label='1w',
                          step='day',
                          stepmode='todate'),
                        # Entire scale
                        dict(step='all')
                    ])
            ),
             # Sliding for selecting time window
             rangeslider=dict(visible=True),
             # Type of xaxis
             type='date'),
            yaxis=dict(title=col),
            showlegend=True
        )

        fig = go.Figure(data=[data], layout=layout)
        iplot(fig)
```

```python
plot_ts("ESS_TOTAL")
```
